## Feature 
### Log Transformation

```python
# Log transformations since our data was skewed showned by the Histogram.
df['log_budget']=np.log1p(df['budget'])
df['log_revenue']=np.log1p(df['revenue'])
```

### Feature scaling

```python
#feture scaling
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df['Budget'] =scaler.fit_transform(df[['log_budget']])
```

```python
scaler=MinMaxScaler()
df['Popularity'] =scaler.fit_transform(df[['popularity']])
```

### Frequency Encoding

```python
df_exploded=df.explode('genres')
df_exploded['Genres']=df_exploded['genres'].map(df_exploded['genres'].value_counts())
df_freq=df_exploded.groupby('title')['Genres'].mean().reset_index()
```

```python
df_final=df.merge(df_freq,on='title').drop(columns=['genres'])
df_final.columns
```

```python
df_final.corr()
```

### Spliting,Training and Testing Data

```python
from sklearn.model_selection import train_test_split
x=df_final.drop(columns=['id', 'title', 'vote_average', 'vote_count', 'status', 'release_date',
       'revenue', 'runtime', 'adult', 'backdrop_path', 'budget', 'homepage',
       'imdb_id', 'original_language', 'original_title', 'overview',
       'popularity', 'poster_path', 'tagline', 'production_companies',
       'production_countries', 'spoken_languages', 'keywords', 'release_month','log_budget','log_revenue',])
y=df_final['log_revenue']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y, test_size=0.2,random_state=42)
```

## Fitting Random Forest Regressor

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model=RandomForestRegressor(n_estimators=200,max_depth=10,min_samples_split=5,random_state=42)
model=LinearRegression()
```

```python
#fitting our model
rf_model.fit(X_train,Y_train)
```

```python
#predicting
y_pred=rf_model.predict(X_test)
```

## Model evaluation

```python
#Using mean squared error
rmse=np.sqrt (mean_squared_error(Y_test,y_pred))
rmse
```

```python
#Using r-squared
from sklearn.metrics import r2_score
r_squared=r2_score(Y_test,y_pred)
r_squared
```

```python
#using the resdiuals
#The plot shows a bell_shaped which implies residuals are normally distributed and hence the model is good
sns.histplot(Y_test-y_pred,bins=20,kde=True)
plt.title("Resdiuals")
```
